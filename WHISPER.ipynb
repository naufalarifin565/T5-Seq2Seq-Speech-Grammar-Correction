WHISPER DENGAN INPUT FILE AUDIO

```python
import whisper
from happytransformer import HappyTextToText, TTSettings

# --- STEP 1: Transkripsi suara dengan Whisper ---
whisper_model = whisper.load_model("base")  # bisa diganti "small", "medium", "large"

# Transcribe audio file
result = whisper_model.transcribe("This kind of mistakes is happens when we not focus during the test..mp3")  # pastikan file ada di direktori
transcribed_text = result["text"]
print("üé§ Transcribed:", transcribed_text)

# Format input sesuai model kamu
model_input = "grammar: " + transcribed_text

# --- STEP 2: Grammar Correction menggunakan model kamu ---
happy_tt = HappyTextToText("T5", "saved_model/")

beam_settings = TTSettings(num_beams=5, min_length=1, max_length=100)

# Generate corrected output
output = happy_tt.generate_text(model_input, args=beam_settings)

print("‚úÖ Grammar Corrected:", output.text)
```

WHISPER DENGAN INPUT SUARA LANGSUNG

```python
import whisper
import sounddevice as sd
from scipy.io.wavfile import write
import numpy as np
import time
from happytransformer import HappyTextToText, TTSettings

# --- STEP 1: Rekam Suara dari Mikrofon ---
def record_audio(duration=5, fs=44100, filename="recorded.wav"):
    print("üéôÔ∏è Mulai merekam...")
    recording = sd.rec(int(duration * fs), samplerate=fs, channels=1, dtype='int16')
    sd.wait()
    write(filename, fs, recording)
    print("‚úÖ Rekaman selesai dan disimpan sebagai:", filename)
    return filename

# Rekam audio 5 detik
audio_path = record_audio(duration=5)

# --- STEP 2: Transkripsi suara dengan Whisper ---
whisper_model = whisper.load_model("base")
result = whisper_model.transcribe(audio_path)
transcribed_text = result["text"]
print("üé§ Transcribed:", transcribed_text)

# Format input sesuai model grammar kamu
model_input = "grammar: " + transcribed_text

# --- STEP 3: Grammar Correction menggunakan model kamu ---
happy_tt = HappyTextToText("T5", "saved_model/")
beam_settings = TTSettings(num_beams=5, min_length=1, max_length=100)
output = happy_tt.generate_text(model_input, args=beam_settings)

print("‚úÖ Grammar Corrected:", output.text)
```
